# Reduction operation 

```python
e = torch.randint(size=(2,3), low=0, high=10, dtype=torch.float32)
e
```

```python
# sum 
torch.sum(e)
torch.sum(e,dim=0) # column wise sum
torch.sum(e,dim=1) # row wise sum 
```

```python
# mean 
torch.mean(e)
torch.mean(e, dim=0)
```

```python
torch.median(e)
```

```python
torch.min(e)
torch.max(e)
torch.std(e)
torch.var(e)
```

```python
torch.argmax(e) # sabse bade element ka index
```

```python
torch.argmin(e) # sabse chhote element ka index
```

# Matrix operations 

```python
a = torch.randint(size=(2,3), low=0, high=10)
b = torch.randint(size=(3,2), low=0, high=10)

print(a)
print(b)
```

```python
# matrix multiplication 
torch.matmul(a,b)
```

```python
vec1 = torch.tensor([1,2])
vec2 = torch.tensor([3,4])

# dot product
torch.dot(vec1, vec2)
```

# Inplace operations 

```python
m = torch.rand(2,3)
n = torch.rand(2,3)

print(m)
print(n)
```

```python
m + n 
```

```python
# I want to store the sum in the location of m only 
m.add_(n)
```

```python
m
```

```python
torch.sigmoid(m)
```

```python
m.sigmoid_()
```
