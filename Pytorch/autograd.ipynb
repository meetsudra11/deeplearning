# simple neural network. x --> sigmoid --> y_pred

```python
import torch 

# inputs 
x = torch.tensor(6.7) # input feature (cgpa)
y = torch.tensor(0.0) # target variable (placement hua k nahi)

w = torch.tensor(1.0)
b = torch.tensor(0.0)
```

```python
# binary cross-entropy loss for scalar 

def binary_cross_entropy(prediction, target):
    epsilon = 1e-7
    prediction = torch.clamp(prediction, epsilon, 1.0 - epsilon)
    return -(target * torch.log(prediction) + (1-target))*torch.log(1-prediction)
```

```python
# forward pass 

z = w * x + b
y_pred = torch.sigmoid(z)

# compute binary cross entropy loss
loss = binary_cross_entropy(y_pred, y)
```

```python
dl_dy_pred = (y_pred - y) / (y_pred * (1-y_pred))

dy_pred_dz = y_pred * (1-y_pred)

dz_dw = x
dz_db = 1

dl_dw = dl_dy_pred * dy_pred_dz * dz_dw
dl_db = dl_dy_pred * dy_pred_dz * dz_db
```

```python
print(dl_dw)
print(dz_dw)
```
