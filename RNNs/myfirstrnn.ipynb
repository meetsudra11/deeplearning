# Sentiment analysis in RNN 

```python
import numpy as np 

docs = [
    'go india',
    'india india',
    'hip hip hurray',
    'jeetega bhai jeetega',
    'bharat mata ki jai',
    'kholi kohli',
    'sachin sachin',
    'dhoni dhoni',
    'modi ki jai',
    'inquilab zindabad'
]
```

```python
from tensorflow.keras.preprocessing.text import Tokenizer
tokenizer = Tokenizer(oov_token='meet')
```

```python
tokenizer.fit_on_texts(docs)
```

```python
tokenizer.word_index
```

```python
tokenizer.word_counts # freq counts
```

```python
tokenizer.document_count
```

```python
sequences = tokenizer.texts_to_sequences(docs)
sequences
```

```python
from keras.utils import pad_sequences
sequences = pad_sequences(sequences, padding="post")
```

```python
sequences
```

#### 2. While for training an rnn using embeddings. U further convert integer encoded values into dense vectors, by adding embedding layer 

```python
from keras.datasets import imdb
from keras import Sequential
from keras.layers import Dense, SimpleRNN, Embedding, Flatten
```

```python
model = Sequential()
model.add(Embedding(18, output_dim=2, input_length=4))

model.summary()
```

```python
model.compile('adam', 'accuracy')
pred = model.predict(sequences)
print(pred)
```

```python
from keras.datasets import imdb
from keras import Sequential
from keras.layers import Dense, SimpleRNN, Embedding, Flatten
```

```python
(X_train, y_train), (X_test,y_test) = imdb.load_data()
```

```python
print(X_train.shape)
print(X_test.shape)
print(y_train.shape)
print(y_test.shape)
```

```python
X_train
```

```python
X_train = pad_sequences(X_train, padding='post', maxlen=50)
X_test = pad_sequences(X_test, padding='post',maxlen=50)
```

```python
X_train.shape
```

```python
model = Sequential()
model.add(Embedding(10000,2))
model.add(SimpleRNN(32, return_sequences=False))
model.add(Dense(1,activation='sigmoid'))

model.summary()
```

```python
model.compile(optimizer='adam', loss='binary_crossentropy', metrics=['acc'])
history = model.fit(X_train, y_train, epochs=5, validation_data=(X_test, y_test))
```

# Imdb dataset - Integer encoding training 

```python
from keras.datasets import imdb
from keras import Sequential
from keras.layers import Dense, SimpleRNN, Embedding, Flatten
```

```python
(X_train, y_train), (X_test,y_test) = imdb.load_data()
```

```python
print(X_train.shape)
print(X_test.shape)
print(y_train.shape)
print(y_test.shape)
```

```python
X_train
```

```python
len(X_train[2])
```

```python
X_train = pad_sequences(X_train, padding='post', maxlen=50)
X_test = pad_sequences(X_test, padding='post',maxlen=50)
```

```python
X_train.shape
```

```python
model = Sequential() 

model.add(SimpleRNN(32, input_shape=(50,1), return_sequences=False))
model.add(Dense(1,activation='sigmoid'))
model.summary()
```

```python
model.compile(loss='binary_crossentropy', optimizer='adam', metrics=['accuracy'])
model.fit(X_train, y_train, epochs=5, validation_data=(X_test,y_test))
```
